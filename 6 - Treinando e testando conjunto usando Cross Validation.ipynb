```python
#matplotlib inline
# -*- coding: cp1252 -*-
import numpy as np
from sklearn.model_selection import KFold #Importando a funo K-fold que vai gerar o iterador
import matplotlib.pyplot as plt

def create(N, VAL):
    #Gerando valores de X e D ("dadosmapalogistico.m")
    xold = np.random.rand()
    x = np.zeros(N)
    
    for k in range(0,N):
        xnew = 4*xold*(1-xold)
        xold = xnew
        x[k] = xnew
    
    x2 = np.transpose(x + VAL*np.std(x)*np.random.randn(1,len(x)))
    x3 = x2[999:(len(x2)-1)] #Descartando valores transitrios
    
    X = x3[0:len(x)-1]
    D = x3[1:len(x)]
    O = [x, x2, x3]
    plot_graphs(O)
    
    return [X, D];

def plot_graphs(O):
    for f in range(0,len(O)):        
        plt.figure(f+1)
        plt.plot(O[f][0:(len(O[f])-1)], O[f][1:len(O[f])], '.')
        plt.xlabel('x(k-1)')
        plt.ylabel('x(k)')
        plt.show()
        
    return;

[X, D] = create(3000, 0.75)      #Sigma = 0.75 definido pelo professor.
```

```python
COMP = [50, 100, 500, 1000] #Comprimentos sugeridos para o conjunto de treinamento
P = 10 #numero de neurnios
N = 5 #Nmero de repeties por neurnios
```

```python
#Utilizaremos uma funo para treinamento da ELM de acordo com o explicado no documento 3
def treinaelm(X,D,P):
    U = np.random.randn(P, 2) #Distribuio Gaussiana
    Xa = np.c_[X, np.ones(X.shape[0])]
    S = Xa@np.transpose(U)
    H = np.tanh(S)
    Ha = np.c_[H,np.ones(X.shape[0])]
    V = np.transpose(np.linalg.pinv(Ha)@D)
    R = Ha@np.transpose(V)
    Y = R
    E = D-Y
    EQM = 1/(D.shape[0])*(np.transpose(E)@E)  #EQM da rede
    return (U,V,Y,E,EQM)
```

```python
#Utilizaremos uma funo para clculo do erro da ELM de acordo com o explicado no documento 3
def erroelm(X,D,U,V):
    Xa = np.c_[X,np.ones(X.shape[0])]
    S = Xa@np.transpose(U)
    H = np.tanh(S)
    Ha = np.c_[H,np.ones(X.shape[0])]
    R = Ha@np.transpose(V)
    Y = R
    E = D-Y
    EQM = 1/(D.shape[0])*(np.transpose(E)@E)
    return (Y,E,EQM)
```

```python
#Funo para realizar o cross-validation para determinada quantidade de P neuronios por N repeties        
def cross_train(N, P, Xtr, Xt, Dtr, Dt):
    # Cross-validation Leave-One-Out
    kf = KFold(n_splits=Xtr.shape[0]) #Definindo o iterador de acordo com o valor de K escolhido
    #Utilizando o iterador criado, e repetindo o processo N vezes para cada P neurnios na camada oculta temos:
    veteqmtr = np.array([])
    veteqmt = np.array([])
    erropredtr = np.ones((N,P))
    erropredt = np.ones((N,P))
    for p in range (1,P+1): #Quantidade de Neurnios ocultos
        for n in range(0,N): #Repeties por neurnio
            for train_index, test_index in kf.split(Xtr):
                X_train, X_test = Xtr[train_index], Xt[test_index]
                D_train, D_test = Dtr[train_index], Dt[test_index]
                [U,V,Ytr,Etr,EQMtr] = treinaelm(X_train,D_train,p)
                veteqmtr = np.append(veteqmtr, EQMtr);
                [Yt,Et,EQMt] = erroelm(X_test,D_test,U,V)
                veteqmt = np.append(veteqmt, EQMt);
            erropredtr[n][p-1] =  np.mean(veteqmtr);
            erropredt[n][p-1] = np.mean(veteqmt);
            veteqmt = []
            veteqmtr = []
    return (erropredtr, erropredt)
```

```python
#Conjunto de treinamento com diferentes tamanhos sendo treinados
cont_figure = 3

for comp in COMP:
    cont_figure+=1
    Xtr = X[0:comp]
    Dtr = D[0:comp]
    Xt = X[comp:2*comp]
    Dt = D[comp:2*comp]
    [erropredtr,erropredt] = cross_train(N, P, Xtr, Xt, Dtr, Dt)

    #Erro de predio no treinamento
    plt.figure(cont_figure)
    plt.title("Treinamento com "+ str(comp)+ " elementos")
    plt.xlabel("Nmero de Neurnios na Camada Oculta")
    plt.ylabel("Erro Quadrtico Mdio")
    plt.boxplot(erropredtr)
    plt.show()

    cont_figure+=1
    
    #Erro de predio no teste
    plt.figure(cont_figure)
    plt.title("Teste com " + str(comp) + " elementos")
    plt.xlabel("Nmero de Neurnios na Camada Oculta")
    plt.ylabel("Erro Quadrtico Mdio")
    plt.boxplot(erropredt)
    plt.show()
```
